```python
import pandas as pd 

```

```python
df = pd.read_csv("../data/csv/game_sales.csv")
df = df.drop(['Name'],axis = 1)
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(14, 6))

# Create the first subplot (left heatmap)
plt.subplot(1, 2, 1)
sns.heatmap(data=df[["count", "Global_Sales"]].corr(), cmap = "viridis",annot=True, alpha=0.5)
plt.title("Correlation between reddits discussion counts and Game sold Globally")
plt.xlabel("discussion counts")
plt.ylabel("Game Sales(Millions)")

# Create the second subplot (right heatmap)
plt.subplot(1, 2, 2)
sns.heatmap(data=df[["count", "NA_Sales"]].corr(), annot=True, alpha=0.5)
plt.title("Correlation between reddits discussion counts and Game sold in North America")
plt.xlabel("discussion counts")
plt.ylabel("Game Sales(Millions)")

# Adjust spacing between subplots
plt.tight_layout()
plt.savefig("../Graphs/yixiang_correlation.jpeg")
```

```python
# Calculate correlations between "count" and "NA_Sales" and "count" and "Global_Sales"
correlation_count_na_sales = df["count"].corr(df["NA_Sales"])
correlation_count_global_sales = df["count"].corr(df["Global_Sales"])

# Create a dedicated line graph for "count" vs. "NA_Sales"
plt.figure(figsize=(11, 8))
plt.scatter(df["count"], df["NA_Sales"], marker='o', color='green',alpha= 0.5,s=50)
plt.title("North America Games sales(Millions) VS Reddit discussion counts")
plt.xlabel("reddit discussion counts")
plt.ylabel("Game sold in North America (Millions)")

# Label specific points with names
points_to_label = df[(df["count"] > 100000) | (df['NA_Sales'] > 4)]
for index, row in points_to_label.iterrows():
    plt.text(row['count'],row['NA_Sales'], row['subreddit'], fontsize=10, ha='left')
plt.savefig("../Graphs/yixiang_game_sales1.jpeg")
```

```python
# Calculate correlations between "count"  and "Global_Sales"
correlation_count_na_sales = df["count"].corr(df["Global_Sales"])
correlation_count_global_sales = df["count"].corr(df["Global_Sales"])

# Create a dedicated line graph for "count" vs. "Global_Sales"
plt.figure(figsize=(11, 8))
plt.scatter(df["count"], df["Global_Sales"], marker='o', color='pink',alpha= 0.5,s=80)
plt.title("Global Games sales(Millions) VS Reddit discussion counts")
plt.xlabel("reddit discussion counts")
plt.ylabel("Game sold globally (Millions)")

# Label specific points with names
points_to_label = df[(df["count"] > 100000) | (df['Global_Sales'] > 5.5)]
for index, row in points_to_label.iterrows():
    plt.text(row['count'],row['Global_Sales'], row['subreddit'], fontsize=13, ha='left')
plt.savefig("../Graphs/yixiang_game_sales2.jpeg")
```
