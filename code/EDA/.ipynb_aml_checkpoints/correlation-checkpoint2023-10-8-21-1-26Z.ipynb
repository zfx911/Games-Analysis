```python
import pandas as pd 

```

```python
df = pd.read_csv("../../data/csv/game_sales.csv")
#df = df.drop(['Name'],axis = 1)
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(14, 6))

# Create the first subplot (left heatmap)
plt.subplot(1, 2, 1)
sns.heatmap(data=df[["count", "Total Sales"]].corr(), cmap = "viridis",annot=True, alpha=0.7)
plt.title("Correlation between reddits discussion counts and Game sold Globally")
plt.xlabel("discussion counts")
plt.ylabel("Game Sales(Millions)")

# Create the second subplot (right heatmap)
plt.subplot(1, 2, 2)
sns.heatmap(data=df[["count", "Critic Score","User Score"]].corr(), annot=True, alpha=0.7)
plt.title("Correlation between reddits discussion counts and Game sold in North America")
plt.xlabel("discussion counts")
plt.ylabel("Game Sales(Millions)")

# Adjust spacing between subplots
plt.tight_layout()
plt.savefig("../../img/correlation.jpeg")
```

```python


# Create a dedicated line graph for "count" vs. "NA_Sales"
plt.figure(figsize=(11, 8))
plt.scatter(df["count"], df["Total Sales"], marker='o', color='green',alpha= 0.5,s=50)
plt.title("Total Games sales(Millions) VS Reddit discussion counts")
plt.xlabel("reddit discussion counts")
plt.ylabel("Game sold worldwied")

# Label specific points with names
points_to_label = df[(df["count"] > 120000) | (df['Total Sales'] >= 8000000)]
for index, row in points_to_label.iterrows():
    plt.text(row['count'],row['Total Sales'], row['subreddit'], fontsize=10, ha='left')
plt.savefig("../../img/yixiang_game_sales1.jpeg")
```

```python


# Create a dedicated line graph for "count" vs. "Global_Sales"
plt.figure(figsize=(11, 8))
plt.scatter(df["count"], df["User Score"], marker='o', color='red',alpha= 0.5,s=80,label='User Score')
plt.scatter(df["count"], df["Critic Score"], marker='o', color='blue',alpha= 0.5,s=80,label='Critics Score')
plt.title("Global Games sales(Millions) VS Reddit discussion counts")
plt.xlabel("reddit discussion counts")
plt.ylabel("Game Scores")
plt.legend()
# Label specific points with names
points_to_label = df[(df["count"] > 100000) | (df['Critic Score'] >= 8.5)]
for index, row in points_to_label.iterrows():
    plt.text(row['count'],row['Critic Score'], row['subreddit'], fontsize=8, ha='left')
plt.savefig("../../img/yixiang_game_sales2.jpeg")
```
